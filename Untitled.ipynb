## ANN Deep Learning | Model Building 

```python
# Checking tensflow for system
import tensorflow as tf
print("Num GPUs Available: ", len(tf.config.experimental.list_physical_devices('GPU')))
```

```python
# Libraries for ANN Deep Learning 
import math
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import Model
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from tensorflow.keras.losses import MeanSquaredLogarithmicError
from tensorflow.keras.initializers import he_normal
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```python
input_feature_size = X_train.shape[1]  # Number of input features after preprocessing
output_feature_size = 1  # Assuming house price prediction is a regression task

# Define fractions for hidden units
hidden_units1_fraction = 0.5  # 50% of input feature size
hidden_units2_fraction = 0.4  # 40% of input feature size
hidden_units3_fraction = 0.3  # 30% of input feature size

# Calculate the actual number of hidden units based on fractions
hidden_units1 = int(hidden_units1_fraction * input_feature_size)
hidden_units2 = int(hidden_units2_fraction * input_feature_size)
hidden_units3 = int(hidden_units3_fraction * input_feature_size)

# Make sure the number of hidden units is not too small or too large
min_hidden_units = 64
max_hidden_units = 512
hidden_units1 = max(min(hidden_units1, max_hidden_units), min_hidden_units)
hidden_units2 = max(min(hidden_units2, max_hidden_units), min_hidden_units)
hidden_units3 = max(min(hidden_units3, max_hidden_units), min_hidden_units)
```

```python
from keras.optimizers import Adam
from keras.regularizers import l2
from keras.initializers import he_normal

def build_model():
    model = Sequential([
        Dense(512, kernel_initializer='he_normal', activation='relu', input_shape=(X_train.shape[1],)),
        BatchNormalization(),
        Dropout(0.5),
        Dense(256, kernel_initializer='he_normal', activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.4),
        Dense(128, kernel_initializer='he_normal', activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dropout(0.3),
        Dense(64, kernel_initializer='he_normal', activation='relu', kernel_regularizer=l2(0.001)),
        BatchNormalization(),
        Dense(1, activation='linear')  # Use 'linear' activation function for regression
    ])

    optimizer = Adam(learning_rate=0.001)
    model.compile(loss='mean_squared_error', optimizer=optimizer)  # Change loss function to 'mean_squared_error'
    return model
```

```python
# build the model
model = build_model()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# loss function
model.compile(
    loss='binary_crossentropy',
    optimizer=Adam(learning_rate=0.01), 
    metrics=['accuracy']
)
# train the model
history = model.fit(
    X_train, 
    y_train, 
    epochs=100, 
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping]
)
```

```python
# Assuming you have already compiled and trained the model with the given code

# Evaluate the model on the test data
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print(f"Test Loss: {test_loss:.4f}")
# test_accuracy = test_accuracy*100
print(f"Test Accuracy: {test_accuracy:.4f}")
```
